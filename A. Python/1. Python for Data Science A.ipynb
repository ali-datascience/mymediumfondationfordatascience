##  Sequence Types
List (list): An ordered, mutable collection of elements.<br>
Tuple (tuple): An ordered, immutable collection of elements.<br>
Range (range): Represents a sequence of numbers.

```python
fruits = ["apple", "banana", "cherry"]  # list
coordinates = (10, 20, 30)               # tuple
numbers = range(1, 5)                    # range
```

```python
# Accessing elements by index
print(fruits[0])

# Modifying elements by index
fruits[1] = "blueberry"
print(fruits)

# Appending an element
fruits.append("orange")
print(fruits)

# Removing an element
fruits.remove("cherry")
print(fruits)

# Slicing the list
slice_fruits = fruits[1:3]
print(slice_fruits)
```

```python
# Accessing elements by index
print(coordinates[0])

# Slicing the tuple
slice_coordinates = coordinates[1:]
print(slice_coordinates)
coordinates[1] = 25
```

```python
# Converting range to a list
range_list = list(numbers)
print(range_list)

# Converting range to a tuple
range_tuple = tuple(numbers)
print(range_tuple)

# Accessing elements by index (after conversion to list)
print(range_list[2])

# Slicing the range (after converting to list)
slice_range = range_list[1:3]
print(slice_range)
```

## Set Types
Set (set): A collection of unique, unordered elements.<br>
Frozenset (frozenset): An immutable version of a set.

```python
colors = {"red", "green", "blue"}     # set
frozen_colors = frozenset({"red", "green"})  # frozenset
```

```python
# Adding an element
colors.add("yellow")
print(colors)

# Removing an element (removes without error if element exists)
colors.remove("green")
print(colors)

# Checking membership (using 'in')
print("red" in colors)
print("green" in colors)

# Set operations (Union, Intersection, Difference)
set1 = {"apple", "banana", "cherry"}
set2 = {"banana", "cherry", "date"}

# Union (combine both sets, removing duplicates)
union_set = set1.union(set2)
print(union_set)

# Intersection (elements common to both sets)
intersection_set = set1.intersection(set2)
print(intersection_set)

# Difference (elements in set1 but not in set2)
difference_set = set1.difference(set2)
print(difference_set)
```

```python
# Trying to remove an element that does not exist (raises KeyError)
colors.remove("pink")
```

```python
# Checking membership (using 'in')
print("red" in frozen_colors)
print("blue" in frozen_colors)

# Set operations (Union, Intersection, Difference)
frozenset1 = frozenset({"apple", "banana", "cherry"})
frozenset2 = frozenset({"banana", "cherry", "date"})

# Union (combine both frozensets, removing duplicates)
union_frozenset = frozenset1.union(frozenset2)
print(union_frozenset)

# Intersection (elements common to both frozensets)
intersection_frozenset = frozenset1.intersection(frozenset2)
print(intersection_frozenset)

# Difference (elements in frozenset1 but not in frozenset2)
difference_frozenset = frozenset1.difference(frozenset2)
print(difference_frozenset)
```

## Binary Types
bytes: Immutable sequence of bytes.<br>
bytearray: Mutable sequence of bytes.<br>
memoryview: A view object that exposes an arrayâ€™s buffer interface.


```python
b = b"hello"          # bytes
ba = bytearray([65, 66, 67])  # bytearray
```

```python
# Accessing elements by index
print(b[0])  # Output: 104 (ASCII value of 'h')

# Slicing the byte sequence
print(b[1:4])  # Output: b'ell'

# Searching for a substring (finding the index of 'l')
index_of_l = b.find(b"l")
print(index_of_l)  # Output: 2 (index of first occurrence of 'l')

# Converting bytes to a string
string_b = b.decode("utf-8")
print(string_b)  # Output: 'hello'
```

```python
b.append(65)
```

```python
# Accessing elements by index
print(ba[0])

# Modifying elements (bytearray is mutable)
ba[0] = 72  # Change 'A' (65) to 'H' (72)
print(ba)  # Output: bytearray(b'HBC')

# Appending a new element
ba.append(68)  # Append 'D' (ASCII value 68)
print(ba)  # Output: bytearray(b'HBCD')

# Slicing the bytearray
print(ba[1:3])  # Output: bytearray(b'BC')

# Converting bytearray to a string
string_ba = ba.decode("utf-8")
print(string_ba)  # Output: 'HBCD'
```

# Operations on Lists

## Creating a List

```python
my_list = [1, 2, 3, 4, 5]
```

## Accessing Elements

```python
print(my_list[0])
print(my_list[2])
print(my_list[-1])
```

## Modifying Elements

```python
my_list[2] = 99
print(my_list)
```

## Adding Elements

```python
my_list.append(6)
print(my_list)

# Using insert() to add an element at a specific index
my_list.insert(2, 77)
print(my_list)

# Using extend() to add multiple elements
my_list.extend([7, 8, 9])
print(my_list)
```

## Removing Elements

```python
my_list.remove(77)
print(my_list)

# Using pop() to remove the last element
last_element = my_list.pop()
print(last_element)
print(my_list)

# Using pop() with an index to remove an element at a specific position
removed_element = my_list.pop(2)
print(removed_element)
print(my_list)

# Using clear() to remove all elements
my_list.clear()
print(my_list)
```

## List Methods
Here are some additional common list methods:

index(): Returns the index of the first occurrence of a value.<br>
count(): Returns the count of a specific value in the list.<br>
sort(): Sorts the list in ascending order.<br>
reverse(): Reverses the list in place.<br>
copy(): Returns a shallow copy of the list

```python
my_list = [3, 1, 4, 1, 5, 9, 2, 6, 5]

# Using index() to find the index of a value
index_of_value = my_list.index(5)
print(index_of_value)

# Using count() to count occurrences of a value
count_of_value = my_list.count(1)
print(count_of_value)

# Using sort() to sort the list
my_list.sort()
print(my_list)

# Using reverse() to reverse the list
my_list.reverse()
print(my_list)

# Using copy() to make a shallow copy of the list
my_list_copy = my_list.copy()
print(my_list_copy)
```

## Slicing Lists

```python
my_list = [3, 1, 4, 1, 5, 9, 2, 6, 5]

print(my_list[1:4])  # (elements from index 1 to 3)
print(my_list[:3])   # (first 3 elements)
print(my_list[3:])   # (elements from index 3 to end)
print(my_list[-3:])  # (last 3 elements)
```

# Operations on Dictionary

## Creating a Dictionary

```python
my_dict = {"name": "Alice", "age": 25, "city": "New York"}

another_dict = dict(name="Bob", age=30, city="Los Angeles")
```

## Accessing Values

```python
print(my_dict["name"])
print(my_dict["age"])

# Using the get() method (returns None if key is not found)
print(my_dict.get("city"))
print(my_dict.get("country"))
```

## Modifying Values

```python
my_dict["age"] = 26
print(my_dict)
```

## Adding Key-Value Pairs

```python
my_dict["country"] = "USA"
print(my_dict)
```

## Removing Key-Value Pairs
You can remove items from a dictionary using different methods:

del: Removes a key-value pair.<br>
pop(): Removes a key-value pair and returns the value.<br>
popitem(): Removes and returns an arbitrary key-value pair.<br>
clear(): Removes all items in the dictionary

```python
# Using del to remove a key-value pair
del my_dict["city"]
print(my_dict)

# Using pop() to remove a specific key-value pair and return the value
removed_value = my_dict.pop("age")
print(removed_value)
print(my_dict)

# Using popitem() to remove and return an arbitrary key-value pair
removed_item = my_dict.popitem()
print(removed_item)
print(my_dict)

# Using clear() to remove all items from the dictionary
my_dict.clear()
print(my_dict)
```

## Dictionary Methods
Here are some common methods available for dictionaries:

keys(): Returns a view object that displays a list of all the keys in the dictionary.<br>
values(): Returns a view object that displays a list of all the values in the dictionary.<br>
items(): Returns a view object that displays a list of all the key-value pairs (tuples) in the dictionary.<br>
update(): Updates the dictionary with elements from another dictionary or an iterable of key-value pairs.<br>
copy(): Returns a shallow copy of the dictionary.

```python
my_dict = {"name": "Alice", "age": 25, "city": "New York"}

# Using keys() to get a view of the dictionary's keys
print(my_dict.keys())

# Using values() to get a view of the dictionary's values
print(my_dict.values())

# Using items() to get a view of the dictionary's key-value pairs
print(my_dict.items())

# Using update() to add or update key-value pairs from another dictionary
another_dict = {"age": 30, "city": "Chicago"}
my_dict.update(another_dict)
print(my_dict)

# Using copy() to create a shallow copy of the dictionary
my_dict_copy = my_dict.copy()
print(my_dict_copy)
```

## Slicing a Dictionary
Unlike lists, dictionaries do not support slicing directly. However, you can convert the dictionary to a list of key-value pairs and slice it if needed

```python
# Converting dictionary to a list of items
dict_items = list(my_dict.items())
print(dict_items)

# Slicing the list of items
sliced_items = dict_items[1:]
print(sliced_items)
```
