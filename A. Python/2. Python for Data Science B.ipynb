# Arithmetic Operators

```python
# Defining two numbers
a = 10
b = 3

# 1. Addition
result_add = a + b
print("Addition:", result_add)

# 2. Subtraction
result_sub = a - b
print("Subtraction:", result_sub)

# 3. Multiplication
result_mul = a * b
print("Multiplication:", result_mul)

# 4. Division
result_div = a / b
print("Division:", result_div)

# 5. Floor Division
result_floor_div = a // b
print("Floor Division:", result_floor_div)

# 6. Modulus
result_mod = a % b
print("Modulus:", result_mod)

# 7. Exponentiation
result_exp = a ** b
print("Exponentiation:", result_exp)
```

```python
# Without parentheses, multiplication and division are evaluated first
result = 10 + 2 * 3
print(result)  # Output: 16 (because 2 * 3 is evaluated first)

# With parentheses, the addition is evaluated first
result_with_parentheses = (10 + 2) * 3
print(result_with_parentheses)  # Output: 36 (because (10 + 2) is evaluated first)
```

# Relational Operators


```python
# Defining variables for comparison
a = 10
b = 20

# 1. Equal to (==)
print(a == b)

# 2. Not equal to (!=)
print(a != b)

# 3. Greater than (>)
print(a > b)

# 4. Less than (<)
print(a < b)

# 5. Greater than or equal to (>=)
print(a >= b)

# 6. Less than or equal to (<=)
print(a <= b)
```

# Assignment Operators

## Basic Assignment (=)

```python
x = 5
```

## Add and Assign (+=)

```python
x = 5
x += 3
print(x)
```

## Subtract and Assign (-=)

```python
x = 5
x -= 2
print(x)
```

## Multiply and Assign (*=)

```python
x = 4
x *= 2
print(x)
```

## Divide and Assign (/=)

```python
x = 10
x /= 2
print(x)
```

## Floor Divide and Assign (//=)

This operator performs floor division (division that rounds down the result) and assigns the result to the left operand.

```python
x = 10
x //= 3
print(x)
```

## Modulus and Assign (%=)
This operator performs modulus (returns the remainder of the division) and assigns the result to the left operand.

```python
x = 10
x %= 3
print(x)
```

##  Exponentiate and Assign (**=)
This operator raises the left operand to the power of the right operand and assigns the result to the left operand.

```python
x = 2
x **= 3
print(x)
```

# Logical Operators

```python
# Defining variables for testing logical operators
a = True
b = False
x = 10
y = 20

# 1. Logical AND (and)
result_and = a and b
print("a and b:", result_and)

# 2. Logical OR (or)
result_or = a or b  # True or False
print("a or b:", result_or)

# 3. Logical NOT (not)
result_not = not a
print("not a:", result_not)

# 4. Using Logical AND with comparison
result_and_comparison = (x > 5) and (y < 30)
print("x > 5 and y < 30:", result_and_comparison)

# 5. Using Logical OR with comparison
result_or_comparison = (x > 15) or (y < 30)
print("x > 15 or y < 30:", result_or_comparison)
```

# Membership Operators

```python
# 1. Using `in` operator with a List
fruits = ['apple', 'banana', 'cherry', 'orange']
print('banana' in fruits)

# 2. Using `in` operator with a String
text = "Hello, world!"
print('Hello' in text)

# 3. Using `in` operator with a Tuple
numbers = (1, 2, 3, 4, 5)
print(3 in numbers)

# 4. Using `not in` operator with a List
fruits = ['apple', 'banana', 'cherry', 'orange']
print('grape' not in fruits)

# 5. Using `not in` operator with a String
text = "Hello, world!"
print('goodbye' not in text)

# 6. Using `in` operator with a Dictionary (checks keys)
person = {'name': 'John', 'age': 30, 'city': 'New York'}
print('name' in person)

# 7. Using `not in` operator with a Dictionary (checks keys)
print('address' not in person)

# 8. Using `in` operator with a Set
colors = {'red', 'green', 'blue'}
print('blue' in colors)
```

# Identity Operators

```python
# 1. Using `is` operator
x = [1, 2, 3]
y = x  # y is assigned the same reference as x
z = [1, 2, 3]  # z is a new list with the same values, but different memory location

print(x is y)  # Output: True (x and y refer to the same object)
print(x is z)  # Output: False (x and z refer to different objects)

# 2. Using `is not` operator
a = 100
b = 100
print(a is not b)  # Output: False (since small integers are cached, a and b refer to the same object)

a = 1000
b = 1000
print(a is not b)  # Output: True (because large integers are not cached, a and b are different objects)

# 3. Using `is` operator with `None`
x = None
print(x is None)  # Output: True (x is the None object)
```
