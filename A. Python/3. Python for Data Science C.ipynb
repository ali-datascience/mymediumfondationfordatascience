# String Formating

## f-Strings (Python 3.6+)

This is a modern, concise, and readable way to format strings.


```python
name = "Alice"
age = 30
print(f"My name is {name} and I am {age} years old.")
```

```python
print(f"The result of 2 + 2 is {2 + 2}.")
```

## str.format() Method
This older method uses placeholders {} to format strings

```python
name = "Bob"
age = 25
print("My name is {} and I am {} years old.".format(name, age))
```

```python
print("{1} is older than {0}.".format("Alice", "Bob"))
print("My name is {name} and I am {age} years old.".format(name="Charlie", age=40))
```

##  Old-Style % Formatting
This is a legacy method similar to C's printf.

```python
name = "Diana"
age = 22
print("My name is %s and I am %d years old." % (name, age))
```

## Template Strings
This method is useful for safe string substitution, especially in user input scenarios

```python
from string import Template

t = Template("My name is $name and I am $age years old.")
print(t.substitute(name="Eve", age=28))
```

# Control Flow IF Else

```python
x = 10
if x > 5:
    print("x is greater than 5")
```

```python
if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")
```

```python
x = 5
if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5 but less than or equal to 10")
else:
    print("x is 5 or less")
```

```python
x = 15
if x > 10:
    if x % 2 == 0:
        print("x is greater than 10 and even")
    else:
        print("x is greater than 10 and odd")
```

```python
x = 7
if x > 0 and x < 10:
    print("x is a single-digit positive number")
```

```python
x = 20
result = "Even" if x % 2 == 0 else "Odd"
print(result)
```

# Iterative Statements



## For

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
for i in range(5):
    print(i)
```

## While

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
while True:
    print("This will run forever unless stopped.")
    break
```

## Using break, continue, and else in Loops

image.png

## break Statement

```python
for num in range(1, 10):
    if num == 5:
        break
    print(num)
```

## continue Statement

```python
for num in range(1, 6):
    if num == 3:
        continue
    print(num)
```

## pass Statement

```python
for num in range(5):
    if num == 3:
        pass  # Placeholder for future code
    else:
        print(num)
```

## Loop with break, continue, and pass

```python
for num in range(1, 8):
    if num == 3:
        pass  # Placeholder, does nothing
    elif num == 5:
        continue  # Skip this iteration
    elif num == 7:
        break  # Exit the loop
    print(num)
```

# List Comprehension

## Basic Example

```python
squares = [x*2 for x in range(5)]
print(squares)
```

## With a Condition

```python
even_numbers = [x for x in range(10) if x % 2 == 0]
print(even_numbers)
```

## Transform Elements

```python
words = ["hello", "world", "python"]
uppercase_words = [word.upper() for word in words]
print(uppercase_words)
```

## Nested Loops

```python
nested_list = [[1, 2], [3, 4], [5, 6]]
flat_list = [item for sublist in nested_list for item in sublist]
print(flat_list)
```

## Conditional Transformation

```python
numbers = range(10)
result = ["even" if x % 2 == 0 else "odd" for x in numbers]
print(result)
```

# Functions

## Defining a Function

```python
def greet(name):
    return f"Hello, {name}!"

# Call the function
print(greet("Alice"))
```

```python
def say_hello():
    print("Hello!")

say_hello()
```

```python
def add(a, b):
    return a + b

result = add(3, 5)
print(result)
```

```python
def greet(name="Guest"):
    return f"Hello, {name}!"

print(greet("Alice"))
print(greet())
```

```python
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(name="Alice", age=30, city="New York")
```

```python
def square(number):
    return number ** 2

print(square(4))
```

## Lambda (Anonymous) Functions

```python
square = lambda x: x**2
print(square(5))
```

```python
add = lambda a, b: a + b
print(add(3, 7))
```

```python
greet = lambda: "Hello, World!"
print(greet())
```

```python
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, numbers))
print(squares)
```

```python
numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)
```
