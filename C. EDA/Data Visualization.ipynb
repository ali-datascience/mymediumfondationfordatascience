# Matplotlib

## Line Plot

```python
import matplotlib.pyplot as plt

# Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot
plt.plot(x, y, label='Line Plot', color='blue', marker='o')

# Adding title and labels
plt.title("Line Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Show grid and legend
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

## Bar Plot

```python
# Sample Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 2, 5, 8]

# Create a bar plot
plt.bar(categories, values, color='green')

# Adding title and labels
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")

# Display the plot
plt.show()
```

## Histogram

```python
import numpy as np

# Sample Data (normally distributed)
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins=30, color='orange', edgecolor='black')

# Adding title and labels
plt.title("Histogram Example")
plt.xlabel("Data")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

## Scatter Plot

```python
# Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a scatter plot
plt.scatter(x, y, color='red')

# Adding title and labels
plt.title("Scatter Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Display the plot
plt.show()
```

## Pie Chart

```python
# Sample Data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['red', 'blue', 'green', 'yellow'])

# Adding title
plt.title("Pie Chart Example")

# Display the plot
plt.show()
```

## Box Plot

```python
# Sample Data
data = [np.random.randn(100) for _ in range(5)]

# Create a box plot
plt.boxplot(data, vert=True, patch_artist=True)

# Adding title and labels
plt.title("Box Plot Example")
plt.ylabel("Values")

# Display the plot
plt.show()
```

## Heatmap

```python
import seaborn as sns

# Sample Data (random matrix)
data = np.random.rand(10, 10)

# Create a heatmap
sns.heatmap(data, cmap='coolwarm', annot=True)

# Adding title
plt.title("Heatmap Example")

# Display the plot
plt.show()
```

## Subplots

```python
# Sample Data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Line plot on the first subplot
axes[0].plot(x, y1, label="Line 1", color='blue')
axes[0].set_title("First Plot")
axes[0].set_xlabel("X Axis")
axes[0].set_ylabel("Y Axis")
axes[0].grid(True)

# Scatter plot on the second subplot
axes[1].scatter(x, y2, color='red')
axes[1].set_title("Second Plot")
axes[1].set_xlabel("X Axis")
axes[1].set_ylabel("Y Axis")

# Display the plots
plt.tight_layout()
plt.show()
```

## Customizing Plot Appearance

```python
# Customizing a line plot
plt.plot(x, y1, linestyle='--', color='purple', marker='s', markersize=8)

# Adding title and labels
plt.title("Customized Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Display the plot
plt.show()
```

## Saving a Plot

```python
# Create a plot
plt.plot(x, y, label='Line Plot', color='blue', marker='o')

# Save the plot to a file
plt.savefig('line_plot_example.png')

# Display the plot
plt.show()
```

# Plotly

##  Line Plot

```python
import plotly.graph_objects as go

# Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a Line Plot
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines', name='Line Plot'))

# Adding title and labels
fig.update_layout(
    title="Plotly Line Plot Example",
    xaxis_title="X Axis",
    yaxis_title="Y Axis"
)

# Display the plot
fig.show()
```

## Bar Plot

```python
# Sample Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 2, 5, 8]

# Create a Bar Plot
fig = go.Figure(data=go.Bar(x=categories, y=values, marker_color='skyblue'))

# Adding title and labels
fig.update_layout(
    title="Plotly Bar Plot Example",
    xaxis_title="Categories",
    yaxis_title="Values"
)

# Display the plot
fig.show()
```

## Histogram

```python
import numpy as np

# Sample Data
data = np.random.randn(1000)

# Create a Histogram
fig = go.Figure(data=go.Histogram(x=data, nbinsx=30, marker_color='orange'))

# Adding title and labels
fig.update_layout(
    title="Plotly Histogram Example",
    xaxis_title="Data",
    yaxis_title="Frequency"
)

# Display the plot
fig.show()
```

## Scatter Plot

```python
# Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a Scatter Plot
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='markers', marker=dict(color='red')))

# Adding title and labels
fig.update_layout(
    title="Plotly Scatter Plot Example",
    xaxis_title="X Axis",
    yaxis_title="Y Axis"
)

# Display the plot
fig.show()
```

## Box Plot

```python
# Sample Data
data = np.random.randn(100)

# Create a Box Plot
fig = go.Figure(data=go.Box(y=data, boxmean=True, marker_color='lightgreen'))

# Adding title and labels
fig.update_layout(
    title="Plotly Box Plot Example",
    yaxis_title="Values"
)

# Display the plot
fig.show()
```

## Heatmap

```python
# Sample Data (correlation matrix)
data = np.random.rand(10, 10)

# Create a Heatmap
fig = go.Figure(data=go.Heatmap(z=data, colorscale='Viridis'))

# Adding title
fig.update_layout(title="Plotly Heatmap Example")

# Display the plot
fig.show()
```

## Violin Plot

```python
import plotly.express as px

# Sample Data
data = np.random.rand(100)

# Create a Violin Plot
fig = px.violin(data, box=True, points="all", title="Plotly Violin Plot Example")

# Display the plot
fig.show()
```

## Pair Plot (Scatter Matrix)

```python
import plotly.express as px

# Sample Data (Iris dataset)
iris = px.data.iris()

# Create a Pair Plot
fig = px.scatter_matrix(iris, dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"], color="species")

# Display the plot
fig.show()
```

## Count Plot

```python
import plotly.express as px

# Sample Data
tips = px.data.tips()

# Create a Count Plot
fig = px.histogram(tips, x="day", color="sex", barmode="group", title="Plotly Count Plot Example")

# Display the plot
fig.show()
```

## Regression Plot

```python
import plotly.express as px

# Sample Data
x = np.random.rand(100)
y = 2 * x + np.random.randn(100)

# Create a Scatter Plot with a regression line
fig = px.scatter(x=x, y=y, trendline="ols", title="Plotly Regression Plot Example")

# Display the plot
fig.show()
```

## Customizing Plot Appearance

```python
# Sample Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 2, 5, 8]

# Create a Bar Plot with custom styling
fig = go.Figure(data=go.Bar(x=categories, y=values, marker_color='darkblue'))

# Customizing Layout
fig.update_layout(
    title="Customized Plotly Bar Plot",
    title_font=dict(size=20, family='Arial'),
    xaxis=dict(title='Categories', title_font=dict(size=15)),
    yaxis=dict(title='Values', title_font=dict(size=15)),
    plot_bgcolor='lightgrey'
)

# Display the plot
fig.show()
```

## Saving the Plot

```python
# Create a Bar Plot
fig = go.Figure(data=go.Bar(x=categories, y=values, marker_color='skyblue'))

# Save the plot as an HTML file
fig.write_html('plotly_bar_plot.html')

# Display the plot
fig.show()
```
