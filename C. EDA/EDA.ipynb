```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

# Load Dataset

```python
# Load Iris dataset from seaborn
data = sns.load_dataset("iris")

# Preview the dataset
print(data.head())
```

# Univariate Analysis

```python
# Shape of the dataset
print("Shape:", data.shape)

# Data types and null values
print(data.info())

# Statistical summary
print(data.describe())
```

```python
# Distribution of numerical variables
data.hist(bins=15, figsize=(10, 6), color='skyblue', edgecolor='black')
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.tight_layout()
plt.show()

# Countplot for species
sns.countplot(data=data, x='species', palette='coolwarm')
plt.title("Count of Each Species")
plt.show()
```

# Bivariate Analysis

```python
# Pairplot for numerical variables
sns.pairplot(data, hue="species", palette="viridis")
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
```

```python
# Correlation matrix
corr_matrix = data.iloc[:, :-1].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Boxplot: Feature distribution by species
for feature in data.columns[:-1]:
    sns.boxplot(data=data, x='species', y=feature, palette="Set3")
    plt.title(f"{feature.capitalize()} Distribution by Species")
    plt.show()
```

#  Multivariate Analysis

## Interactive 3D Scatter Plot

```python
fig = px.scatter_3d(data, x='sepal_length', y='sepal_width', z='petal_length',
                    color='species', size='petal_width',
                    title="3D Scatter Plot of Iris Features",
                    labels={'sepal_length': 'Sepal Length',
                            'sepal_width': 'Sepal Width',
                            'petal_length': 'Petal Length'})
fig.show()
```

# Outlier Detection

```python
# Boxplot for outliers
for col in data.columns[:-1]:
    sns.boxplot(x=data[col], color='cyan')
    plt.title(f"Outliers in {col}")
    plt.show()
```
