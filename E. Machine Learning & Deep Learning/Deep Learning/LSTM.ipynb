```python
# Import necessary libraries
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Generate some dummy sequential data
data = np.random.randn(1000, 10)  # 1000 sequences of length 10
targets = np.random.randint(0, 2, size=(1000,))  # Binary classification targets

# Define LSTM model architecture
model = Sequential()
model.add(LSTM(32, input_shape=(10, 1)))  # 32 LSTM units, input sequences of length 10
model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Reshape the data for LSTM input (samples, timesteps, features)
data = data.reshape((data.shape[0], data.shape[1], 1))

# Train the LSTM model
model.fit(data, targets, epochs=10, batch_size=32)

# Evaluate the model
loss, accuracy = model.evaluate(data, targets)
print(f"Test Loss: {loss}, Test Accuracy: {accuracy}")
```
