```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense


# Generate simple sequential data using np.random
def generate_data(n_samples, time_steps):
    X = np.random.randint(0, 100, size=(n_samples, time_steps, 1))
    y = np.sum(X, axis=1)  # Sum of each sequence
    return X, y

# Parameters
n_samples = 1000  # Number of data samples
time_steps = 10   # Length of input sequence

# Generate data
X, y = generate_data(n_samples, time_steps)

# Split data into training and testing sets
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Build RNN model
model = Sequential([
    SimpleRNN(64, input_shape=(time_steps, 1), activation='relu'),
    Dense(1)
])

# Compile model
model.compile(optimizer='adam', loss='mse')

# Train model
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.1)

# Evaluate model
test_loss = model.evaluate(X_test, y_test)
print("Test Loss:", test_loss)

# Predictions
predictions = model.predict(X_test)

# Printing some predictions for illustration
print("Some Predictions:")
for i in range(5):
    print("True:", y_test[i], "Predicted:", predictions[i][0])
```
